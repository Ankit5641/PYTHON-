# JOINS IN PANDAS

```python
# used to joins two sheet or dataset
# joins => left , right , inner , outer 
# merge 
```

```python
import pandas as pd

employees = pd.DataFrame({
    "emp_id": [101, 102, 103, 104, 105, 106],
    "name": ["Aman", "Riya", "Sahil", "Neha", "Rahul", "Pooja"],
    "dept_id": [1, 2, 1, 3, 4, None],
    "salary": [60000, 45000, 55000, 75000, 50000, 48000]
})

employees
```

```python
departments = pd.DataFrame({
    "dept_id": [1, 2, 3, 5],
    "department": ["IT", "HR", "Finance", "Marketing"]
})

departments
```

```python
locations = pd.DataFrame({
    "dept_id": [1, 2, 3],
    "location": ["Bangalore", "Delhi", "Mumbai"]
})

locations
```

```python
# inner join 
```

```python
pd.merge(employees, departments, on="dept_id", how="inner")
```

```python
# left join 
pd.merge(employees, departments, on="dept_id", how="left")
```

```python
# right join
pd.merge(employees, departments, on="dept_id", how="right")
```

```python
# outer join 
pd.merge(employees, departments, on="dept_id", how="outer")
```

```python
df = pd.merge(employees, departments, on="dept_id", how="left")
df = pd.merge(df, locations, on="dept_id", how="left")
df
```

```python
departments_indexed = departments.set_index("dept_id")
employees.join(departments_indexed, on="dept_id")
```
